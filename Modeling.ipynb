```python
plt.figure(figsize=(14, 8))
plt.style.use('seaborn-v0_8-whitegrid')

plt.plot(population_data['year'], population_data['total_population_k'], 
         linewidth=3, color='#2E86AB', marker='o', markersize=4)

plt.title('Динамика общего населения', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Год', fontsize=12, fontweight='bold')
plt.ylabel('Население, тыс. чел.', fontsize=12, fontweight='bold')

plt.gca().get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, p: format(int(x), ','))
)

plt.grid(True, alpha=0.3)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.style.use('seaborn-v0_8-whitegrid')

plt.plot(population_data['year'], population_data['male_population_k'], 
         linewidth=2.5, color='#1f77b4', label='Мужчины', marker='s', markersize=3)
plt.plot(population_data['year'], population_data['female_population_k'], 
         linewidth=2.5, color='#ff7f0e', label='Женщины', marker='^', markersize=3)

plt.title('Динамика населения по полу', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Год', fontsize=12, fontweight='bold')
plt.ylabel('Население, тыс. чел.', fontsize=12, fontweight='bold')

plt.gca().get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, p: format(int(x), ','))
)

plt.legend(frameon=True, fancybox=True, shadow=True, fontsize=12)

plt.grid(True, alpha=0.3)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
alive_males = people_data[(people_data['IsAlive'] == True) & 
                        (people_data['Gender'] == 'Male')]


bins = [0, 18, 45, 65, 100]
labels = ['0-18', '19-45', '46-65', '66-100']
alive_males['AgeGroup'] = pandas.cut(alive_males['Age'], bins=bins, labels=labels, right=True)

age_distribution = alive_males['AgeGroup'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
bars = plt.bar(age_distribution.index, age_distribution.values, 
               color='lightblue', edgecolor='darkblue', alpha=0.7)

plt.title('Возрастной состав мужского населения на конец моделирования', fontsize=14)
plt.xlabel('Возрастные группы', fontsize=12)
plt.ylabel('Количество (тыс. чел.)', fontsize=12)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', 
             ha='center', va='bottom', fontweight='bold')

plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
